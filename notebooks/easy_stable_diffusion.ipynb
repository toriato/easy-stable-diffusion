```python
#@markdown ## ì›í´ë¦­ ì½”ë©
import io
import json
import os
import shlex
import shutil
import signal
import subprocess
import sys
import time
from datetime import datetime
from distutils.spawn import find_executable
from importlib.util import find_spec
from pathlib import Path
from typing import Callable, List, Optional, Tuple, Union

import requests

OPTIONS = {}

# fmt: off
#####################################################
# ì½”ë© ë…¸íŠ¸ë¶ì—ì„  #@param ë¬¸ë²•ìœ¼ë¡œ ì‚¬ìš©ìë¡œë¶€í„° ì„¤ì • ê°’ì„ ê°€ì ¸ì˜¬ ìˆ˜ ìˆìŒ
# ë‹¤ë¥¸ í™˜ê²½ì¼ ë• override.json íŒŒì¼ ë“±ì„ ì‚¬ìš©í•´ì•¼í•¨
#####################################################
#@title

#@markdown ##### <font color="orange">***Torch ë²„ì „ ì„ íƒ***</font>
TORCH_VERSION = "torch==1.13.1+cu117"  # @param ["torch==1.13.1+cu117", "torch==2.0.0+cu118", "ê¸°ë³¸ ë²„ì „"]
#@markdown - <font color="green">ì„ íƒ A</font>: torch==1.13.1+cu117, ddetailer í™•ì¥ ì‚¬ìš© ê°€ëŠ¥
#@markdown - <font color="red">ì„ íƒ B</font>: torch==2.0.0+cu118, ddetailer í™•ì¥ ì‚¬ìš© ë¶ˆê°€

#@markdown ##### <font color="orange">***ddetailer ì˜ì¡´ íŒ¨í‚¤ì§€ë¥¼ ë¯¸ë¦¬ ì„¤ì¹˜í• ì§€?***</font>
#@markdown ì•„ë˜ íŒ¨í‚¤ì§€ë¥¼ ë¯¸ë¦¬ ì„¤ì¹˜í•´ mmcv-full ë²„ì „ ë¬¸ì œë¡œ ì¸í•œ í™•ì¥ ì„¤ì¹˜ ë¬¸ì œë¥¼ í•´ê²° í•©ë‹ˆë‹¤.<br> ì„¤ì¹˜ íŒ¨í‚¤ì§€ : openmim==0.3.7, mmcv-full==1.7.1, mmdet==2.28.2
#@markdown - <font color="green">ì²´í¬ì‹œ</font>: ddetailer í™•ì¥ì„ ì‚¬ìš©í•˜ëŠ” ê²½ìš° í•„ìš”í•œ íŒ¨í‚¤ì§€ë¥¼ ë¯¸ë¦¬ ì„¤ì¹˜
#@markdown - <font color="red">í•´ì œì‹œ</font>: ddetailer í™•ì¥ì„ ì‚¬ìš©í•˜ì§€ ì•Šìœ¼ë©´ ì²´í¬ í•´ì œ
INSTALL_DDETAILER_REQUIREMENTS = True  #@param {type:"boolean"}
OPTIONS['INSTALL_DDETAILER_REQUIREMENTS'] = INSTALL_DDETAILER_REQUIREMENTS

#@markdown ##### <font color="orange">***ì‘ì—… ë””ë ‰í„°ë¦¬ ê²½ë¡œ***</font>
#@markdown ì„ë² ë”©, ëª¨ë¸, ê²°ê³¼ì™€ ì„¤ì • íŒŒì¼ ë“±ì´ ì˜êµ¬ì ìœ¼ë¡œ ë³´ê´€ë  ë””ë ‰í„°ë¦¬ ê²½ë¡œ
WORKSPACE = 'SD' #@param {type:"string"}

#@markdown ##### <font color="orange">***ìë™ìœ¼ë¡œ ì½”ë© ëŸ°íƒ€ì„ì„ ì¢…ë£Œí• ì§€?***</font>
DISCONNECT_RUNTIME = True  #@param {type:"boolean"}
OPTIONS['DISCONNECT_RUNTIME'] = DISCONNECT_RUNTIME

#@markdown ##### <font color="orange">***êµ¬ê¸€ ë“œë¼ì´ë¸Œì™€ ë™ê¸°í™”í• ì§€?***</font>
#@markdown <font color="red">**ì£¼ì˜**</font>: ë™ê¸°í™” ì „ ë‚¨ì€ ìš©ëŸ‰ì´ ì¶©ë¶„í•œì§€ í™•ì¸ í•„ìˆ˜ (5GB ì´ìƒ)
USE_GOOGLE_DRIVE = True  #@param {type:"boolean"}
OPTIONS['USE_GOOGLE_DRIVE'] = USE_GOOGLE_DRIVE

#@markdown ##### <font color="orange">***xformers ë¥¼ ì‚¬ìš©í• ì§€?***</font>
#@markdown ì„ íƒí•œ Torch ë²„ì „ì— ë”°ë¼ 0.0.16rc425(torch==1.13.1), 0.0.17(torch==2.0.0)ì´ ì„¤ì¹˜
#@markdown - <font color="green">ì¥ì </font>: ì´ë¯¸ì§€ ìƒì„± ì†ë„ ê°œì„  ê°€ëŠ¥ì„± ìˆìŒ
#@markdown - <font color="red">ë‹¨ì </font>: ì¶œë ¥í•œ ê·¸ë¦¼ì˜ ì§ˆì´ ì¡°ê¸ˆ ë–¨ì–´ì§ˆ ìˆ˜ ìˆìŒ
USE_XFORMERS = True  #@param {type:"boolean"}
OPTIONS['USE_XFORMERS'] = USE_XFORMERS

#@markdown ##### <font color="orange">***ì¸ì¦ ì •ë³´***</font>
#@markdown ì ‘ì† ì‹œ ì‚¬ìš©í•  ì‚¬ìš©ì ì•„ì´ë””ì™€ ë¹„ë°€ë²ˆí˜¸
#@markdown <br>`GRADIO_USERNAME` ì…ë ¥ ë€ì— `user1:pass1,user,pass2`ì²˜ëŸ¼ ì…ë ¥í•˜ë©´ ì—¬ëŸ¬ ì‚¬ìš©ì ì¶”ê°€ ê°€ëŠ¥
#@markdown <br>`GRADIO_USERNAME` ì…ë ¥ ë€ì„ <font color="red">ë¹„ì›Œë‘ë©´</font> ì¸ì¦ ê³¼ì •ì„ ì‚¬ìš©í•˜ì§€ ì•ŠìŒ
GRADIO_USERNAME = '' #@param {type:"string"}
GRADIO_PASSWORD = '' #@param {type:"string"}
OPTIONS['GRADIO_USERNAME'] = GRADIO_USERNAME
OPTIONS['GRADIO_PASSWORD'] = GRADIO_PASSWORD

#@markdown ##### <font color="orange">***í„°ë„ë§ ì„œë¹„ìŠ¤***</font>
TUNNEL = 'gradio' #@param ["none", "gradio", "cloudflared", "ngrok"]
TUNNEL_URL: Optional[str] = None
OPTIONS['TUNNEL'] = TUNNEL

#@markdown ##### <font color="orange">***ngrok API í‚¤***</font>
#@markdown ngrok í„°ë„ì— ì‚¬ìš©í•  API í† í°
#@markdown <br>[ì„¤ì •í•˜ëŠ” ë°©ë²•ì€ ì—¬ê¸°ë¥¼ í´ë¦­í•´ í™•ì¸](https://arca.live/b/aiart/60683088), [API í† í°ì€ ì—¬ê¸°ë¥¼ ëˆŒëŸ¬ ê³„ì •ì„ ë§Œë“  ë’¤ ì–»ì„ ìˆ˜ ìˆìŒ](https://dashboard.ngrok.com/get-started/your-authtoken)
#@markdown <br>ì…ë ¥ ë€ì„ <font color="red">ë¹„ì›Œë‘ë©´</font> ngrok í„°ë„ì„ ë¹„í™œì„±í™”í•¨
#@markdown - <font color="green">ì¥ì </font>: ì ‘ì†ì´ ë¹ ë¥¸ í¸ì´ê³  íƒ€ì„ì•„ì›ƒì´ ê±°ì˜ ë°œìƒí•˜ì§€ ì•ŠìŒ
#@markdown - <font color="red">**ë‹¨ì **</font>: ê³„ì •ì„ ë§Œë“¤ê³  API í† í°ì„ ì§ì ‘ ì…ë ¥í•´ì¤˜ì•¼í•¨
NGROK_API_TOKEN = '' #@param {type:"string"}
OPTIONS['NGROK_API_TOKEN'] = NGROK_API_TOKEN

#@markdown ##### <font color="orange">***WebUI ë ˆí¬ì§€í† ë¦¬ ì£¼ì†Œ***</font>
REPO_URL = 'https://github.com/AUTOMATIC1111/stable-diffusion-webui.git' #@param {type:"string"}
OPTIONS['REPO_URL'] = REPO_URL

#@markdown ##### <font color="orange">***WebUI ë ˆí¬ì§€í† ë¦¬ ì»¤ë°‹ í•´ì‹œ***</font>
#@markdown ì—…ë°ì´íŠ¸ê°€ ì‹¤ì‹œê°„ìœ¼ë¡œ ì˜¬ë¼ì˜¬ ë•Œ ìµœì‹  ë²„ì „ì—ì„œ ì˜¤ë¥˜ê°€ ë°œìƒí•  ë•Œ [ë ˆí¬ì§€í† ë¦¬ ì»¤ë°‹ ëª©ë¡](https://github.com/AUTOMATIC1111/stable-diffusion-webui/commits/master)ì—ì„œ
#@markdown <br>ê³¼ê±° ì»¤ë°‹ í•´ì‹œ ê°’[(ì˜ë¬¸ê³¼ ìˆ«ìë¡œëœ ë‚œìˆ˜ ê°’; ì˜ˆì‹œ ì´ë¯¸ì§€)](https://vmm.pw/MzMy)ì„ ì•„ë˜ì— ë¶™ì—¬ë„£ì€ ë’¤ ì‹¤í–‰í•˜ë©´ ê³¼ê±° ë²„ì „ì„ ì‚¬ìš©í•  ìˆ˜ ìˆìŒ
#@markdown <br>ì…ë ¥ ë€ì„ <font color="red">ë¹„ì›Œë‘ë©´</font> ê°€ì¥ ìµœì‹  ì»¤ë°‹ì„ ê°€ì ¸ì˜´
REPO_COMMIT = '' #@param {type:"string"}
OPTIONS['REPO_COMMIT'] = REPO_COMMIT

#@markdown ##### <font color="orange">***Python ë°”ì´ë„ˆë¦¬ ì´ë¦„***</font>
#@markdown ì…ë ¥ ë€ì„ <font color="red">ë¹„ì›Œë‘ë©´</font> ì‹œìŠ¤í…œì— ì„¤ì¹˜ëœ Python ì„ ì‚¬ìš©í•¨
PYTHON_EXECUTABLE = '' #@param {type:"string"}
OPTIONS['PYTHON_EXECUTABLE'] = PYTHON_EXECUTABLE

#@markdown ##### <font color="orange">***WebUI ì¸ì***</font>
#@markdown <font color="red">**ì£¼ì˜**</font>: ë¹„ì–´ìˆì§€ ì•Šìœ¼ë©´ ì‹¤í–‰ì— í•„ìš”í•œ ì¸ìê°€ ìë™ìœ¼ë¡œ ìƒì„±ë˜ì§€ ì•ŠìŒ
#@markdown <br>[ì‚¬ìš©í•  ìˆ˜ ìˆëŠ” ì¸ì ëª©ë¡](https://github.com/AUTOMATIC1111/stable-diffusion-webui/blob/master/modules/shared.py#L23)
ARGS = '' #@param {type:"string"}
OPTIONS['ARGS'] = shlex.split(ARGS)

#@markdown ##### <font color="orange">***WebUI ì¶”ê°€ ì¸ì***</font>
EXTRA_ARGS = '' #@param {type:"string"}
OPTIONS['EXTRA_ARGS'] = shlex.split(EXTRA_ARGS)

#####################################################
# ì‚¬ìš©ì ì„¤ì • ê°’ ë
#####################################################
# fmt: on

# ë¡œê·¸ ë³€ìˆ˜
LOG_FILE: Optional[io.TextIOWrapper] = None
LOG_WIDGET = None
LOG_BLOCKS = []

# ë¡œê·¸ HTML ìœ„ì ¯ ìŠ¤íƒ€ì¼
LOG_WIDGET_STYLES = {
    'wrapper': {
        'overflow-x': 'auto',
        'max-width': '100%',
        'padding': '1em',
        'background-color': 'black',
        'white-space': 'pre',
        'font-family': 'monospace',
        'font-size': '1em',
        'line-height': '1.1em',
        'color': 'white'
    },
    'dialog': {
        'display': 'block',
        'margin-top': '.5em',
        'padding': '.5em',
        'font-weight': 'bold',
        'font-size': '1.5em',
        'line-height': '1em',
        'color': 'black'
    }
}
LOG_WIDGET_STYLES['dialog_success'] = {
    **LOG_WIDGET_STYLES['dialog'],
    'border': '3px dashed darkgreen',
    'background-color': 'green',
}
LOG_WIDGET_STYLES['dialog_error'] = {
    **LOG_WIDGET_STYLES['dialog'],
    'border': '3px dashed darkred',
    'background-color': 'red',
}

IN_INTERACTIVE = hasattr(sys, 'ps1')
IN_COLAB = False

try:
    from IPython import get_ipython
    IN_COLAB = 'google.colab' in str(get_ipython())
except ImportError:
    pass


def hook_runtime_disconnect():
    """
    ì…€ì´ ì¢…ë£Œëì„ ë•Œ ìë™ìœ¼ë¡œ ëŸ°íƒ€ì„ì„ í•´ì œí•˜ë„ë¡ asyncio ìŠ¤ë ˆë“œë¥¼ ìƒì„±í•©ë‹ˆë‹¤
    """
    if not IN_COLAB:
        return

    from google.colab import runtime

    # asyncio ëŠ” ì—¬ëŸ¬ ê²¹ìœ¼ë¡œ ì‚¬ìš©í•  ìˆ˜ ì—†ê²Œë” ì„¤ê³„ëê¸° ë•Œë¬¸ì—
    # ì£¼í”¼í„° ë…¸íŠ¸ë¶ ë“± ì´ë¯¸ ë£¨í”„ê°€ ëŒê³  ìˆëŠ” ê³³ì—ì„  ì‚¬ìš©í•  ìˆ˜ ì—†ìŒ
    # ì´ëŠ” nest-asyncio íŒ¨í‚¤ì§€ë¥¼ í†µí•´ ì–´ëŠì •ë„ ìš°íšŒí•˜ì—¬ ì‚¬ìš©í•  ìˆ˜ ìˆìŒ
    # https://pypi.org/project/nest-asyncio/
    if not has_python_package('nest_asyncio'):
        execute(['pip', 'install', 'nest-asyncio'])

    import nest_asyncio
    nest_asyncio.apply()

    import asyncio

    async def unassign():
        time.sleep(1)
        runtime.unassign()

    # í‰ë²”í•œ í™˜ê²½ì—ì„  ë¹„ë™ê¸°ë¡œ ë™ì‘í•˜ì—¬ ë°”ë¡œ ì‹¤í–‰ë˜ë‚˜
    # ì½”ë© ëŸ°íƒ€ì„ì—ì„  ìˆœì°¨ì ìœ¼ë¡œ ì‹¤í–‰ë˜ê¸° ë•Œë¬¸ì— í˜„ì¬ ì…€ ì¢…ë£Œ í›„ ì¦‰ì‹œ ì‹¤í–‰ë¨
    asyncio.create_task(unassign())


def setup_tunnels():
    global TUNNEL_URL

    tunnel = OPTIONS['TUNNEL']

    if tunnel == 'none':
        pass

    elif tunnel == 'gradio':
        if not has_python_package('gradio'):
            # https://fastapi.tiangolo.com/release-notes/#0910
            execute(['pip', 'install', 'gradio', 'fastapi==0.90.1'])

        import secrets

        from gradio.networking import setup_tunnel
        TUNNEL_URL = setup_tunnel('localhost', 7860, secrets.token_urlsafe(32))

    elif tunnel == 'cloudflared':
        if not has_python_package('pycloudflared'):
            execute(['pip', 'install', 'pycloudflared'])

        from pycloudflared import try_cloudflare
        TUNNEL_URL = try_cloudflare(port=7860).tunnel

    elif tunnel == 'ngrok':
        if not has_python_package('pyngrok'):
            execute(['pip', 'install', 'pyngrok'])

        auth = None
        token = OPTIONS['NGROK_API_TOKEN']

        if ':' in token:
            parts = token.split(':')
            auth = parts[1] + ':' + parts[-1]
            token = parts[0]

        from pyngrok import conf, exception, ngrok, process

        # ë¡œì»¬ í¬íŠ¸ê°€ ë‹«í˜€ìˆìœ¼ë©´ ê²½ê³  ë©”ì„¸ì§€ê°€ ìŠ¤íŒ¸ë§ˆëƒ¥ ì¶œë ¥ë˜ë¯€ë¡œ ì˜¤ë¥˜ë§Œ í‘œì‹œë˜ê²Œ ìˆ˜ì •í•¨
        process.ngrok_logger.setLevel('ERROR')

        try:
            tunnel = ngrok.connect(
                7860,
                pyngrok_config=conf.PyngrokConfig(
                    auth_token=token,
                    region='jp'
                ),
                auth=auth,
                bind_tls=True
            )
        except exception.PyngrokNgrokError:
            alert('ngrok ì—°ê²°ì— ì‹¤íŒ¨í–ˆìŠµë‹ˆë‹¤, í† í°ì„ í™•ì¸í•´ì£¼ì„¸ìš”!', True)
        else:
            assert isinstance(tunnel, ngrok.NgrokTunnel)
            TUNNEL_URL = tunnel.public_url

    else:
        raise ValueError(f'{tunnel} ì— ëŒ€ì‘í•˜ëŠ” í„°ë„ ì„œë¹„ìŠ¤ê°€ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤')


def setup_environment():
    # ë…¸íŠ¸ë¶ í™˜ê²½ì´ë¼ë©´ ë¡œê·¸ í‘œì‹œë¥¼ ìœ„í•œ HTML ìš”ì†Œ ë§Œë“¤ê¸°
    if IN_INTERACTIVE:
        try:
            from IPython.display import display
            from ipywidgets import widgets

            global LOG_WIDGET
            LOG_WIDGET = widgets.HTML()
            display(LOG_WIDGET)

        except ImportError:
            pass

    # êµ¬ê¸€ ë“œë¼ì´ë¸Œ ë§ˆìš´íŠ¸í•˜ê¸°
    if IN_COLAB and OPTIONS['USE_GOOGLE_DRIVE']:
        from google.colab import drive
        drive.mount('/content/drive')

        global WORKSPACE
        WORKSPACE = str(
            Path('drive', 'MyDrive', WORKSPACE).resolve()
        )

    # ë¡œê·¸ íŒŒì¼ ë§Œë“¤ê¸°
    global LOG_FILE
    workspace = Path(WORKSPACE).resolve()
    log_path = workspace.joinpath(
        'logs',
        datetime.strftime(datetime.now(), '%Y-%m-%d_%H-%M-%S.log')
    )

    log_path.parent.mkdir(0o777, True, True)

    LOG_FILE = log_path.open('a')

    # í˜„ì¬ í™˜ê²½ ì¶œë ¥
    import platform
    log(' '.join(os.uname()))
    log(f'Python {platform.python_version()}')
    log(str(Path().resolve()))

    # ë®ì–´ì“¸ ì„¤ì • íŒŒì¼ ê°€ì ¸ì˜¤ê¸°
    override_path = workspace.joinpath('override.json')
    if override_path.exists():
        with override_path.open('r') as file:
            override_options = json.loads(file.read())
            for key, value in override_options.items():
                if key not in OPTIONS:
                    log(f'{key} í‚¤ëŠ” ì¡´ì¬í•˜ì§€ ì•ŠëŠ” ì„¤ì •ì…ë‹ˆë‹¤', styles={'color': 'red'})
                    continue

                if type(value) != type(OPTIONS[key]):
                    log(f'{key} í‚¤ëŠ” {type(OPTIONS[key]).__name__} ìë£Œí˜•ì´ì—¬ë§Œ í•©ë‹ˆë‹¤', styles={
                        'color': 'red'})
                    continue

                OPTIONS[key] = value

                log(f'override.json: {key} = {json.dumps(value)}')

    if IN_COLAB:
        # ë‹¤ë¥¸ Python ë²„ì „ ì„¤ì¹˜
        if OPTIONS['PYTHON_EXECUTABLE'] and not find_executable(OPTIONS['PYTHON_EXECUTABLE']):
            execute(['apt', 'install', OPTIONS['PYTHON_EXECUTABLE']])
            execute(
                f"curl -sS https://bootstrap.pypa.io/get-pip.py | {OPTIONS['PYTHON_EXECUTABLE']}"
            )

        # ì„ íƒí•œ í† ì¹˜ ë²„ì „ ì„¤ì¹˜
        if 'torch==1.13.1+cu117' in TORCH_VERSION:
          execute(['pip', 'install', '-q', '-U', 'torch==1.13.1+cu117', 'torchvision==0.14.1+cu117', 'torchaudio==0.13.1+cu117', 'torchtext==0.14.1', 'torchdata==0.5.1', '--extra-index-url', 'https://download.pytorch.org/whl/cu117'])

          if OPTIONS['USE_XFORMERS']:
            execute(['pip', 'install', '-q', '-U', 'xformers==0.0.16rc425'])
          
          # ddetailer ì˜ì¡´ íŒ¨í‚¤ì§€ ë¯¸ë¦¬ ì„¤ì¹˜
          if INSTALL_DDETAILER_REQUIREMENTS:
            execute(['pip', 'install', '-q', '-U', 'openmim==0.3.7'])
            execute(['mim', 'install', '-q', '-U', 'mmcv-full==1.7.1'])
            execute(['pip', 'install', '-q', '-U', 'mmdet==2.28.2'])

        elif 'torch==2.0.0+cu118' in TORCH_VERSION:
          execute(['pip', 'install', '-q', '-U', 'torch==2.0.0+cu118', 'torchvision==0.15.1+cu118', 'torchaudio==2.0.1+cu118', 'torchtext==0.15.1', 'torchdata==0.6.0', '--extra-index-url', 'https://download.pytorch.org/whl/cu118'])

          if OPTIONS['USE_XFORMERS']:
            execute(['pip', 'install', '-q', '-U', 'xformers==0.0.17'])

        # ëŸ°íƒ€ì„ì´ ì •ìƒì ìœ¼ë¡œ ì´ˆê¸°í™” ëëŠ”ì§€ í™•ì¸í•˜ê¸°
        try:
            import torch
        except:
            alert('torch íŒ¨í‚¤ì§€ê°€ ì˜ëª»ëìŠµë‹ˆë‹¤, ëŸ°íƒ€ì„ì„ ë‹¤ì‹œ ì‹¤í–‰í•´ì£¼ì„¸ìš”!', True)
        else:
            if not torch.cuda.is_available():
                alert('GPU ëŸ°íƒ€ì„ì´ ì•„ë‹™ë‹ˆë‹¤, í• ë‹¹ëŸ‰ì´ ì´ˆê³¼ ëì„ ìˆ˜ë„ ìˆìŠµë‹ˆë‹¤!')

                OPTIONS['EXTRA_ARGS'] += [
                    '--skip-torch-cuda-test',
                    '--no-half',
                    '--opt-sub-quad-attention'
                ]

    # ì™¸ë¶€ í„°ë„ë§ ì´ˆê¸°í™”
    setup_tunnels()

    # ì²´í¬í¬ì¸íŠ¸ ëª¨ë¸ì´ ì¡´ì¬í•˜ì§€ ì•ŠëŠ”ë‹¤ë©´ ê¸°ë³¸ ëª¨ë¸ ë°›ì•„ì˜¤ê¸°
    if not has_checkpoint():
        for file in [
            {
                'url': 'https://huggingface.co/gsdf/Counterfeit-V2.5/resolve/main/Counterfeit-V2.5_fp16.safetensors',
                'target': str(workspace.joinpath('models/Stable-diffusion/Counterfeit-V2.5_fp16.safetensors')),
                'summary': 'ê¸°ë³¸ ì²´í¬í¬ì¸íŠ¸ íŒŒì¼ì„ ë°›ì•„ì˜µë‹ˆë‹¤'
            },
            {
                'url': 'https://huggingface.co/saltacc/wd-1-4-anime/resolve/main/VAE/kl-f8-anime2.ckpt',
                'target': str(workspace.joinpath('models/VAE/kl-f8-anime2.ckpt')),
                'summary': 'ê¸°ë³¸ VAE íŒŒì¼ì„ ë°›ì•„ì˜µë‹ˆë‹¤'
            }
        ]:
            download(**file)


# ==============================
# ë¡œê·¸
# ==============================


def format_styles(styles: dict) -> str:
    return ';'.join(map(lambda kv: ':'.join(kv), styles.items()))


def render_log() -> None:
    try:
        from ipywidgets import widgets
    except ImportError:
        return

    if not isinstance(LOG_WIDGET, widgets.HTML):
        return

    html = f'''<div style="{format_styles(LOG_WIDGET_STYLES['wrapper'])}">'''

    for block in LOG_BLOCKS:
        styles = {
            'display': 'inline-block',
            **block['styles']
        }
        child_styles = {
            'display': 'inline-block',
            **block['child_styles']
        }

        html += f'<span style="{format_styles(styles)}">{block["msg"]}</span>\n'

        if block['max_childs'] is not None and len(block['childs']) > 0:
            html += f'<div style="{format_styles(child_styles)}">'
            html += ''.join(block['childs'][-block['max_childs']:])
            html += '</div>'

    html += '</div>'

    LOG_WIDGET.value = html


def log(
    msg: str,
    styles={},
    newline=True,

    parent=False,
    parent_index: Optional[int] = None,
    child_styles={
        'padding-left': '1em',
        'color': 'gray'
    },
    max_childs=0,

    print_to_file=True,
    print_to_widget=True
) -> Optional[int]:
    # ê¸°ë¡í•  ë‚´ìš©ì´ ngrok API í‚¤ì™€ ì¼ì¹˜í•œë‹¤ë©´ ìˆ¨ê¸°ê¸°
    # TODO: ë” ë‚˜ì€ ë¬¸ìì—´ ê²€ì‚¬, ì›ì¹˜ ì•Šì€ ë‚´ìš©ì´ ê°€ë ¤ì§ˆ ìˆ˜ë„ ìˆìŒ
    if OPTIONS['NGROK_API_TOKEN'] != '':
        msg = msg.replace(OPTIONS['NGROK_API_TOKEN'], '**REDACTED**')

    if newline:
        msg += '\n'

    # íŒŒì¼ì— ê¸°ë¡í•˜ê¸°
    if print_to_file and LOG_FILE:
        if parent_index and msg.endswith('\n'):
            LOG_FILE.write('\t')
        LOG_FILE.write(msg)
        LOG_FILE.flush()

    # ë¡œê·¸ ìœ„ì ¯ì— ê¸°ë¡í•˜ê¸°
    if print_to_widget and LOG_WIDGET:
        # ë¶€ëª¨ ë¡œê·¸ê°€ ì—†ë‹¤ë©´ ìƒˆ ë¸”ë¡ ë§Œë“¤ê¸°
        if parent or parent_index is None:
            LOG_BLOCKS.append({
                'msg': msg,
                'styles': styles,
                'childs': [],
                'child_styles': child_styles,
                'max_childs': max_childs
            })
            render_log()
            return len(LOG_BLOCKS) - 1

        # ë¶€ëª¨ ë¡œê·¸ê°€ ì¡´ì¬í•œë‹¤ë©´ ì¶”ê°€í•˜ê¸°
        if len(LOG_BLOCKS[parent_index]['childs']) > 100:
            LOG_BLOCKS[parent_index]['childs'].pop(0)

        LOG_BLOCKS[parent_index]['childs'].append(msg)
        render_log()

    print('\t' if parent_index else '' + msg, end='')


def log_trace() -> None:
    import sys
    import traceback

    # ìŠ¤íƒ ê°€ì ¸ì˜¤ê¸°
    ex_type, ex_value, ex_traceback = sys.exc_info()

    styles = {}

    # ì˜¤ë¥˜ê°€ ì¡´ì¬í•œë‹¤ë©´ ë©”ì„¸ì§€ ë¹¨ê°„ìƒ‰ìœ¼ë¡œ ì¶œë ¥í•˜ê¸°
    # https://docs.python.org/3/library/sys.html#sys.exc_info
    # TODO: ì˜¤ë¥˜ ìœ ë¬´ ì´ë ‡ê²Œ í™•ì¸í•˜ë©´ ì•ˆë ê±° ê°™ì€ë° ì¼ë‹¨ ê·€ì°®ì•„ì„œ ëŒ€ì¶© ì¨ë‘ 
    if ex_type is not None:
        styles = LOG_WIDGET_STYLES['dialog_error']

    parent_index = log(
        'ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤, <a href="https://discord.gg/6wQeA2QXgM">ë””ìŠ¤ì½”ë“œ ì„œë²„</a>ì— ë³´ê³ í•´ì£¼ì„¸ìš”',
        styles)
    assert parent_index

    # ì˜¤ë¥˜ê°€ ì¡´ì¬í•œë‹¤ë©´ ì˜¤ë¥˜ ì •ë³´ì™€ ìŠ¤íƒ íŠ¸ë ˆì´ìŠ¤ ì¶œë ¥í•˜ê¸°
    if ex_type is not None:
        log(f'{ex_type.__name__}: {ex_value}', parent_index=parent_index)
        log(
            '\n'.join(traceback.format_tb(ex_traceback)),
            parent_index=parent_index
        )

    # ë¡œê·¸ íŒŒì¼ì´ ì—†ìœ¼ë©´ ë³´ê³ í•˜ì§€ ì•Šê¸°
    # TODO: ë¡œê·¸ íŒŒì¼ì´ ì¡´ì¬í•˜ì§€ ì•Šì„ ìˆ˜ê°€ ìˆë‚˜...?
    if not LOG_FILE:
        log('ë¡œê·¸ íŒŒì¼ì´ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤, ë³´ê³ ì„œë¥¼ ë§Œë“¤ì§€ ì•ŠìŠµë‹ˆë‹¤')
        return


def alert(message: str, unassign=False):
    log(message)

    if IN_INTERACTIVE:
        from IPython.display import display
        from ipywidgets import widgets

        display(
            widgets.HTML(f'<script>alert({json.dumps(message)})</script>')
        )

    if IN_COLAB and unassign:
        from google.colab import runtime

        time.sleep(1)
        runtime.unassign()


# ==============================
# ì„œë¸Œ í”„ë¡œì„¸ìŠ¤
# ==============================
def execute(
    args: Union[str, List[str]],
    parser: Optional[
        Callable[[str], None]
    ] = None,
    summary: Optional[str] = None,
    hide_summary=False,
    print_to_file=True,
    print_to_widget=True,
    **kwargs
) -> Tuple[str, int]:
    if isinstance(args, str) and 'shell' not in kwargs:
        kwargs['shell'] = True

    # ì„œë¸Œ í”„ë¡œì„¸ìŠ¤ ë§Œë“¤ê¸°
    p = subprocess.Popen(
        args,
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        encoding='utf-8',
        **kwargs)

    # ë¡œê·¸ì— ì‹œì‘í•œ í”„ë¡œì„¸ìŠ¤ ì •ë³´ ì¶œë ¥í•˜ê¸°
    formatted_args = args if isinstance(args, str) else ' '.join(args)
    summary = formatted_args if summary is None else f'{summary}\n   {formatted_args}'

    log_index = log(
        f'=> {summary}',
        styles={'color': 'yellow'},
        max_childs=10)

    output = ''

    # í”„ë¡œì„¸ìŠ¤ ì¶œë ¥ ìœ„ì ¯ì— ë¦¬ë‹¤ì´ë ‰ì…˜í•˜ê¸°
    while p.poll() is None:
        # ì¶œë ¥ì´ ë¹„ì–´ìˆë‹¤ë©´ ë„˜ì–´ê°€ê¸°
        assert p.stdout
        line = p.stdout.readline()
        if not line:
            continue

        # í”„ë¡œì„¸ìŠ¤ ì¶œë ¥ ë²„í¼ì— ì¶”ê°€í•˜ê¸°
        output += line

        # íŒŒì„œ í•¨ìˆ˜ ì‹¤í–‰í•˜ê¸°
        if callable(parser):
            parser(line)

        # í”„ë¡œì„¸ìŠ¤ ì¶œë ¥ ë¡œê·¸í•˜ê¸°
        log(
            line,
            newline=False,
            parent_index=log_index,
            print_to_file=print_to_file,
            print_to_widget=print_to_widget)

    # ë³€ìˆ˜ ì •ë¦¬í•˜ê¸°
    rc = p.poll()
    assert rc is not None

    # ë¡œê·¸ ë¸”ë¡ ì—…ë°ì´íŠ¸
    if LOG_WIDGET:
        assert log_index

        if rc == 0:
            # í˜„ì¬ ë¡œê·¸ í…ìŠ¤íŠ¸ ì´ˆë¡ìƒ‰ìœ¼ë¡œ ë³€ê²½í•˜ê³  í”„ë¡œì„¸ìŠ¤ ì¶œë ¥ ìˆ¨ê¸°ê¸°
            LOG_BLOCKS[log_index]['styles']['color'] = 'green'
            LOG_BLOCKS[log_index]['max_childs'] = None
        else:
            # í˜„ì¬ ë¡œê·¸ í…ìŠ¤íŠ¸ ë¹¨ê°„ìƒ‰ìœ¼ë¡œ ë³€ê²½í•˜ê³  í”„ë¡œì„¸ìŠ¤ ì¶œë ¥ ëª¨ë‘ í‘œì‹œí•˜ê¸°
            LOG_BLOCKS[log_index]['styles']['color'] = 'red'
            LOG_BLOCKS[log_index]['max_childs'] = 0

        if hide_summary:
            # í˜„ì¬ ë¡œê·¸ ë¸”ë¡ ìˆ¨ê¸°ê¸° (ì œê±°í•˜ê¸°)
            del LOG_BLOCKS[log_index]

        # ë¡œê·¸ ë¸”ë¡ ë Œë”ë§
        render_log()

    # ì˜¤ë¥˜ ì½”ë“œë¥¼ ë°˜í™˜í–ˆë‹¤ë©´
    if rc != 0:
        if isinstance(rc, signal.Signals):
            rc = rc.value

        raise subprocess.CalledProcessError(rc, args)

    return output, rc

# ==============================
# ì‘ì—… ê²½ë¡œ
# ==============================


def delete(path: os.PathLike) -> None:
    path = Path(path)

    if path.is_file() or path.is_symlink():
        path.unlink()
    else:
        shutil.rmtree(path, ignore_errors=True)


def has_python_package(pkg: str, executable: Optional[str] = None) -> bool:
    if not executable:
        return find_spec(pkg) is not None

    _, rc = execute(
        [
            executable, '-c',
            f'''
            import importlib
            import sys
            sys.exit(0 if importlib.find_loader({shlex.quote(pkg)}) else 0)
            '''
        ])

    return True if rc == 0 else False


# ==============================
# íŒŒì¼ ë‹¤ìš´ë¡œë“œ
# ==============================
def download(url: str, target: Optional[str] = None, ignore_aria2=False, **kwargs):
    if not target:
        # TODO: ê²½ë¡œ ì¤‘ params ì œê±°í•˜ê¸°
        target = url.split('/')[-1]

    # íŒŒì¼ì„ ë°›ì„ ë””ë ‰í„°ë¦¬ ë§Œë“¤ê¸°
    Path(target).parent.mkdir(0o777, True, True)

    # ë¹ ë¥¸ ë‹¤ìš´ë¡œë“œë¥¼ ìœ„í•´ aria2 íŒ¨í‚¤ì§€ ì„¤ì¹˜ ì‹œë„í•˜ê¸°
    if not ignore_aria2:
        if not find_executable('aria2c') and find_executable('apt'):
            execute(['apt', 'install', 'aria2'])

        if find_executable('aria2c'):
            p = Path(target)
            execute(
                [
                    'aria2c',
                    '--continue',
                    '--always-resume',
                    '--summary-interval', '10',
                    '--disk-cache', '64M',
                    '--min-split-size', '8M',
                    '--max-concurrent-downloads', '8',
                    '--max-connection-per-server', '8',
                    '--max-overall-download-limit', '0',
                    '--max-download-limit', '0',
                    '--split', '8',
                    '--dir', str(p.parent),
                    '--out', p.name,
                    url
                ],
                **kwargs)

    elif find_executable('curl'):
        execute(
            [
                'curl',
                '--location',
                '--output', target,
                url
            ],
            **kwargs)

    else:
        if 'summary' in kwargs.keys():
            log(kwargs.pop('summary'), **kwargs)

        with requests.get(url, stream=True) as res:
            res.raise_for_status()

            with open(target, 'wb') as file:
                # ë°›ì•„ì˜¨ íŒŒì¼ ë””ì½”ë”©í•˜ê¸°
                # https://github.com/psf/requests/issues/2155#issuecomment-50771010
                import functools
                res.raw.read = functools.partial(
                    res.raw.read,
                    decode_content=True)

                # TODO: íŒŒì¼ ê¸¸ì´ê°€ ì í•©í•œì§€?
                shutil.copyfileobj(res.raw, file, length=16*1024*1024)


def has_checkpoint() -> bool:
    workspace = Path(WORKSPACE)
    for p in workspace.joinpath('models', 'Stable-diffusion').glob('**/*'):
        if p.suffix != '.ckpt' and p.suffix != '.safetensors':
            continue

        # aria2 ë¡œ ë°›ë‹¤ë§Œ íŒŒì¼ì´ë©´ ë¬´ì‹œí•˜ê¸°
        if p.with_suffix(p.suffix + '.aria2c').exists():
            continue

        return True
    return False


def parse_webui_output(line: str) -> None:
    # ì²« ì‹œì‘ì— í•œí•´ì„œ ì›¹ ì„œë²„ ì—´ë ¸ì„ ë•Œ ë‹¤ì´ì–´ë¡œê·¸ í‘œì‹œí•˜ê¸°
    if 'Running on local URL:' in line:
        log(
            '\n'.join([
                'ì„±ê³µì ìœ¼ë¡œ í„°ë„ì´ ì—´ë ¸ìŠµë‹ˆë‹¤',
                f'<a target="_blank" href="{TUNNEL_URL}">{TUNNEL_URL}</a>',
            ]),
            LOG_WIDGET_STYLES['dialog_success']
        )
        return


def setup_webui() -> None:
    repo_dir = Path('repository')
    need_clone = True

    # ì´ë¯¸ ë””ë ‰í„°ë¦¬ê°€ ì¡´ì¬í•œë‹¤ë©´ ì •ìƒì ì¸ ë ˆí¬ì¸ì§€ í™•ì¸í•˜ê¸°
    if repo_dir.is_dir():
        try:
            # ì‚¬ìš©ì íŒŒì¼ë§Œ ë‚¨ê²¨ë‘ê³  ë ˆí¬ì§€í† ë¦¬ ì´ˆê¸°í™”í•˜ê¸°
            # https://stackoverflow.com/a/12096327
            execute(
                'git stash && git pull origin master',
                cwd=repo_dir
            )
        except subprocess.CalledProcessError:
            log('ë ˆí¬ì§€í† ë¦¬ê°€ ì˜ëª»ëìŠµë‹ˆë‹¤, ë””ë ‰í„°ë¦¬ë¥¼ ì œê±°í•©ë‹ˆë‹¤')
        else:
            need_clone = False

    # ë ˆí¬ì§€í† ë¦¬ í´ë¡ ì´ í•„ìš”í•˜ë‹¤ë©´ ê¸°ì¡´ ë””ë ‰í„°ë¦¬ ì§€ìš°ê³  í´ë¡ í•˜ê¸°
    if need_clone:
        shutil.rmtree(repo_dir, ignore_errors=True)
        execute(['git', 'clone', OPTIONS['REPO_URL'], str(repo_dir)])

    # íŠ¹ì • ì»¤ë°‹ì´ ì§€ì •ëë‹¤ë©´ ì²´í¬ì•„ì›ƒí•˜ê¸°
    if OPTIONS['REPO_COMMIT']:
        execute(
            ['git', 'checkout', OPTIONS['REPO_COMMIT']],
            cwd=repo_dir
        )

    if IN_COLAB:
        patch_path = repo_dir.joinpath('scripts', 'patches.py')

        if not patch_path.exists():
            download(
                'https://raw.githubusercontent.com/toriato/easy-stable-diffusion/main/scripts/patches.py',
                str(patch_path),
                ignore_aria2=True)


def start_webui(args: List[str] = OPTIONS['ARGS']) -> None:
    workspace = Path(WORKSPACE).resolve()
    repository = Path('repository').resolve()

    # ê¸°ë³¸ ì¸ì ë§Œë“¤ê¸°
    if len(args) < 1:
        args += ['--data-dir', str(workspace)]

        # xformers
        if OPTIONS['USE_XFORMERS']:
            try:
                import torch
            except ImportError:
                pass
            else:
                if torch.cuda.is_available():
                    args += [
                        '--xformers',
                        '--xformers-flash-attention'
                    ]

        # Gradio ì¸ì¦ ì •ë³´
        if OPTIONS['GRADIO_USERNAME'] != '':
            args += [
                f'--gradio-auth',
                OPTIONS['GRADIO_USERNAME'] +
                ('' if OPTIONS['GRADIO_PASSWORD'] ==
                    '' else ':' + OPTIONS['GRADIO_PASSWORD'])
            ]

    # ì¶”ê°€ ì¸ì
    args += OPTIONS['EXTRA_ARGS']

    env = {
        **os.environ,
        'HF_HOME': str(workspace / 'cache' / 'huggingface')
    }

    try:
        execute(
            [
                OPTIONS['PYTHON_EXECUTABLE'] or 'python',
                '-u',
                '-m', 'launch',
                *args
            ],
            parser=parse_webui_output,
            cwd=str(repository),
            env=env,
            start_new_session=True,
        )
    except subprocess.CalledProcessError as e:
        if IN_COLAB and e.returncode == signal.SIGINT.value:
            raise RuntimeError(
                'í”„ë¡œì„¸ìŠ¤ê°€ ê°•ì œ ì¢…ë£ŒëìŠµë‹ˆë‹¤, ë©”ëª¨ë¦¬ê°€ ë¶€ì¡±í•´ ë°œìƒí•œ ë¬¸ì œì¼ ìˆ˜ë„ ìˆìŠµë‹ˆë‹¤') from e


try:
    setup_environment()

    # 3ë‹¨ ì´ìƒ(?) ë ˆë²¨ì—ì„œ ì‹¤í–‰í•˜ë©´ nested ëœ asyncio ì´ ë¬¸ì œë¥¼ ì¼ìœ¼í‚´
    # ëŸ°íƒ€ì„ì„ ì¢…ë£Œí•´ë„ ì½”ë© í˜ì´ì§€ì—ì„  ëŸ°íƒ€ì„ì´ ì‹¤í–‰ ì¤‘(Busy)ì¸ ê²ƒìœ¼ë¡œ í‘œì‹œë˜ë¯€ë¡œ ì—¬ê¸°ì„œ ì‹¤í–‰í•¨
    if OPTIONS['DISCONNECT_RUNTIME']:
        hook_runtime_disconnect()

    setup_webui()
    start_webui()

# ^c ì¢…ë£Œ ë¬´ì‹œí•˜ê¸°
except KeyboardInterrupt:
    pass

except:
    # ë¡œê·¸ ìœ„ì ¯ì´ ì—†ë‹¤ë©´ í‰ë²”í•˜ê²Œ ì˜¤ë¥˜ ì²˜ë¦¬í•˜ê¸°
    if not LOG_WIDGET:
        raise

    log_trace()
```
